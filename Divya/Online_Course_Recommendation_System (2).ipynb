### Outlier Detection

```python
num_cols = df.select_dtypes(include=np.number).columns
plt.figure(figsize=(15,15))
for i, col in enumerate(num_cols):
    plt.subplot(3,3,i+1)
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot of {col}')
    plt.ylabel('Values')
plt.tight_layout()
plt.show()
```

```python
def outlier_capping(df,columns):
    Q1= df[columns].quantile(0.25)
    Q3= df[columns].quantile(0.75)
    IQR= Q3-Q1
    lower_extreme= Q1-1.5*IQR
    upper_extreme= Q3+1.5*IQR
    df[columns]=df[columns].apply(lambda x: lower_extreme if x<lower_extreme else upper_extreme if x>upper_extreme else x)
for col in df.select_dtypes(include=['int','float']).columns:
    outlier_capping(df,col)
```

### Model Building
Create a Copy of Dataset

Create a copy of the dataset for model building.

```python
course_df = df.copy()

course_df.head()
```

#### Sampling Dataset

The dataset contains 1 lakh records. To avoid memory issues and improve processing speed, we use a sample of 1000 records.

```python
course_df = course_df.sample(
    1000,
    random_state=42
)

course_df.shape
```

#### Feature Selection

Select important course attributes required for recommendation.

```python
features = [
    'course_name',
    'instructor',
    'difficulty_level',
    'certification_offered',
    'study_material_available'
]

course_df[features].head()
```

#### Missing Value Handling


Check and handle missing values in selected features.

```python
course_df[features].isnull().sum()
```

```python
for feature in features:
    course_df[feature] = course_df[feature].fillna('')
```

#### Feature Engineering


Combine selected features into a single text column.

```python
course_df['combined_features'] = (
    course_df['course_name'].astype(str) + ' ' +
    course_df['instructor'].astype(str) + ' ' +
    course_df['difficulty_level'].astype(str) + ' ' +
    course_df['certification_offered'].astype(str) + ' ' +
    course_df['study_material_available'].astype(str) + ' ' +
    course_df['previous_courses_taken'].astype(str)
)
```

```python
# Remove duplicate courses
course_df = course_df.drop_duplicates(subset='course_name')

# Reset index
course_df = course_df.reset_index(drop=True)

print("New Shape:", course_df.shape)
```

```python
# TF-IDF Vectorization
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(course_df['combined_features'])
```

```python
course_df[
    ['course_name',
     'combined_features']
].head()
```

#### Import Required Libraries

Import TF-IDF Vectorizer and Cosine Similarity libraries.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

#### TF-IDF Vectorization

Convert textual information into numerical vectors.

```python
tfidf = TfidfVectorizer(
    stop_words='english'
)

tfidf_matrix = tfidf.fit_transform(
    course_df['combined_features']
)
```

```python
tfidf_matrix.shape
```

#### Calculate Cosine Similarity

Calculate similarity scores between courses.

```python
cosine_sim = cosine_similarity(
    tfidf_matrix
)
```

```python
cosine_sim.shape
```

#### Reset Dataset Index

Reset index values for easy course retrieval.

```python
course_df = course_df.reset_index(
    drop=True
)
```

#### Build Recommendation Function

Generate top 5 similar course recommendations.

```python
def recommend_courses(course_name):

    try:

        idx = course_df[
            course_df['course_name']
            == course_name
        ].index[0]

        similarity_scores = list(
            enumerate(cosine_sim[idx])
        )

        similarity_scores = sorted(
            similarity_scores,
            key=lambda x:x[1],
            reverse=True
        )

        similarity_scores = similarity_scores[1:6]

        print(
            "Top 5 Recommended Courses:"
        )

        for i in similarity_scores:

            print(
                course_df.iloc[i[0]]
                ['course_name']
            )

    except:

        print("Course not found")
```

#### Create Streamlit Web Application

Create Streamlit Application

```python
%%writefile app.py

import streamlit as st
import pandas as pd
import joblib

# Load saved files
course_df = pd.read_pickle("course_df.pkl")
cosine_sim = joblib.load("cosine_sim.pkl")

# Title
st.title("Online Course Recommendation System")

st.write("Select a course to get recommendations")

# Course list
course_list = course_df['course_name'].unique()

selected_course = st.selectbox(
    "Choose Course",
    course_list
)

# Recommendation Function
def recommend_courses(course_name):

    idx = course_df[
        course_df['course_name'] == course_name
    ].index[0]

    similarity_scores = list(
        enumerate(cosine_sim[idx])
    )

    similarity_scores = sorted(
        similarity_scores,
        key=lambda x:x[1],
        reverse=True
    )

    similarity_scores = similarity_scores[1:6]

    recommendations = []

    for i in similarity_scores:

        recommendations.append(
            course_df.iloc[i[0]]['course_name']
        )

    return recommendations

# Button
if st.button("Recommend Courses"):

    recommendations = recommend_courses(
        selected_course
    )

    st.subheader("Top 5 Recommended Courses")

    for course in recommendations:
        st.write(course)
```
